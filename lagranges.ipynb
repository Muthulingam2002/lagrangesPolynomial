```python
import numpy as np
import matplotlib.pyplot as plt
x=np.random.uniform(-100, 100, 5)
y=np.random.uniform(-100, 100, 5)
x.sort()
xarr=np.linspace(x[0],x[-1])
yarr=np.array([],float)
for xpoint in xarr:
    ypoint = 0
    for i in range(len(x)):
        p = 1
        for j in range(len(x)):
            if i != j:
                p = p * (xpoint - x[j])/(x[i] - x[j])
        ypoint = ypoint + p * y[i]
    yarr=np.append(yarr,ypoint)
plt.title("Lagrange Polynomial")
plt.plot(xarr,yarr,"b-",x,y,"ro")
plt.show()
```
